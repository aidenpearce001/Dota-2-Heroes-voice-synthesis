```python
from bs4 import BeautifulSoup,Tag, NavigableString
import requests
import os
import urllib.request
import re

from unicodedata import normalize
```

```python
link = "https://dota2.gamepedia.com/Monkey_King/Responses"
req = requests.get(link)
soup = BeautifulSoup(req.text, "lxml")
```

```python
path = link.split("/")[-2]
if not os.path.exists(path):
    os.makedirs(link.split("/")[-2])
```

```python
f = open(path+'/data.txt','w+')

TAG_RE = re.compile(r'<[^>]+>')

ul = soup.find_all('ul')
for i,v in enumerate(ul):
    audio = ul[i].find('audio')
    if audio:
        for i in ul[i]:
            if isinstance(i, NavigableString) == False:
                text = TAG_RE.sub('', str(i))
                text = text.replace("Link","")
                print(text)
                ext = path+'/'+i.find('audio').source['src'].split("/")[-1]
                if len(text) != 1 and "arcana" not in i.find('audio').source['src']:
                    f.write(ext.split('.')[-2]+'.wav|'+text.lstrip()+'\n')
                    print(i.find('audio').source['src'])
                urllib.request.urlretrieve(i.find('audio').source['src'], path+'/'+i.find('audio').source['src'].split("/")[-1])
                print("==========================================================================")
f.close()
```

```python
def head(file_name,N):
    with open(file_name) as myfile:
        head = [next(myfile) for x in range(N)]
    return head
```

```python
def tail(file_name, N):
    
    list_of_lines = []
    
    with open(file_name, 'rb') as test_set:
        test_set.seek(0, os.SEEK_END)
        buffer = bytearray()
        pointer_location = test_set.tell()
        while pointer_location >= 0:
            test_set.seek(pointer_location)
            pointer_location = pointer_location -1
            new_byte = test_set.read(1)
            if new_byte == b'\n':
                list_of_lines.append(buffer.decode()[::-1])
                if len(list_of_lines) == N:
                    return list(reversed(list_of_lines))
                buffer = bytearray()
            else:
                buffer.extend(new_byte)
        if len(buffer) > 0:
            list_of_lines.append(buffer.decode()[::-1])
    return list(reversed(list_of_lines))
```

```python
total = 0
with open(path+'/data.txt') as f:
    for i, l in enumerate(f):
        total += 1
total
```

```python
test = round(total * 20 / 100)
train = total - test

test_file = tail(path+'/data.txt', test)
train_file = head(path+'/data.txt',train)

with open('test_files.txt','w+') as f:
    for line in test_file:
        f.write(str(line))
        
with open('train_files.txt','w+') as f:
    for line in train_file:
        f.write(str(line))
```

```python
import glob
import wave
import contextlib
from pydub import AudioSegment

for name in glob.glob(path+'/*.mp3'):
    sound = AudioSegment.from_mp3(name)
    sound.export(name.split('.')[-2]+'.wav', format="wav")
    os.remove(name)
    
time = 0
for name in glob.glob(path+'/*.wav'):
    with contextlib.closing(wave.open(name,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        time += duration
        print(name+':'+str(duration)+' secs')
print('We got total '+str(time/60)+ ' mins of sounds')
    
```
